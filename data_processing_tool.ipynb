# Data Preprocessing Tools

## Importing the libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset


```python
dataset = pd.read_csv('Data.csv')
X = dataset.iloc[:, :-1].values #[:, :-1]=>[rows, column-1]
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

## Taking care of missing data


```python
dataset.Salary.median()
```

```python
dataset.Salary=dataset.Salary.fillna(dataset.Salary.median())
dataset
```

```python
dataset.Age.median()
```

```python
dataset.Age=dataset.Age.fillna(dataset.Age.median())
dataset
```

Or it can be done by using the sklearn libary

```python
from sklearn.impute import SimpleImputer
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(X[:,1:3])
X[:,1:3]=imputer.transform(X[:,1:3])
```

```python
dataset
```

## Encoding categorical data


### Encoding the Independent Variable


```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder


ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough') #[0] python index starts from '0'
X=np.array(ct.fit_transform(X))
```

```python
print(X)
```

### Encoding the Dependent Variable


```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)     #label encoding
```

```python
y
```

## Splitting the dataset into the Training set and Test set


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```
